The groupby function is used to group data based on 1 or more columns in a dataframe.

```python
df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
N_Qty=df.groupby(['Size'],as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
sns.barplot(x='Size',y='Qty',data=N_Qty)
```

Note: According to the above graph, the majority of quantity purchases m-size in sales.

```python
#histogram
df['Size'].hist()
```

```python
df['Category'] = df['Category'].astype(str)
column_data = df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data,bins=30,edgecolor='Black')
#plt.xticks(rotation=90)(if u want to rotate x-axis lables then use this line)
plt.show()
```

```python
#checking B2B data by using pie chart
B2B_check = df['B2B'].value_counts()
```

```python
#plot the pie chart
plt.pie(B2B_check,labels=B2B_check,autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
#plot the pie chart
plt.pie(B2B_check,labels=B2B_check.index,autopct='%1.1f%%')
#plt.axis('equal')
plt.show()
```

```python
Note: From the above graph 99.2% of purchasers are retailers,and 0.8% are(business-to-business(B2B)) buyers.
```

```python
#prepare data for scatter plot
x_data = df['Category']
y_data = df['Size']
```

```python
# plot the scatter plot
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('scatter plot')
plt.show()
```

```python
#plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top 10 states
top_10_state = df['ship-state'].value_counts().head(10)
#plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)],x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of state')
plt.xticks(rotation=90)
plt.show()
```

Note: The graph indicates that the majority of purchasers are from the state of Maharashtra.
